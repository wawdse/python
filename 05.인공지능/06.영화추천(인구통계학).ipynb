```python
df1.rename(columns={'movie_id':'id'}, inplace=True)
df1.head(2)
```

```python
df_temp = df1[['id', 'cast', 'crew']]
df_temp.head(2)
```

```python
df = df2.merge(df_temp, on='id')
df.columns
```

```python
df[['vote_count', 'vote_average']].head()
```

```python
#평점 가중치 구하기
#WR = (v/(v+m) * R) + (m/(v+m) * C)
```

```python
C = df['vote_average'].mean()
C
```

```python
m = df['vote_count'].quantile(0.9)
m 
```

```python
df_movies = df.copy()
filt = df['vote_count'] >= m
df_movies = df_movies[filt]
```

```python
df.shape, df_movies.shape
```

```python
df_movies['vote_count'].sort_values()
```

```python
def weighted_rating(x):
    m = df['vote_count'].quantile(0.9)
    C = df['vote_average'].mean()
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m)* R) + (m/(v+m) * C)
```

```python
df_movies['score'] = df_movies.apply(weighted_rating, axis=1)
```

```python
df_movies[['title', 'score']]
```

```python
df_movies = df_movies.sort_values('score', ascending=False)
df_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```
