```python
code = '005930'
ss = fdr.DataReader(code, '2020-01-01', '2021-08-30')
ss.tail()
```

```python
n_train = int(len(ss) * 0.7)

len(ss), n_train
```

```python
y_train = ss['Close'][:n_train]
y_test = ss['Close'][n_train:]
```

```python
len(y_train), len(y_test)
```

```python
y_train[:5], y_test[:5]
```

```python
plt.figure(figsize=(10,3))
y_train.plot(color="#FFEB35")
y_test.plot(color="#8CFF8C")
plt.show()
```

```python
from pmdarima.arima import ndiffs
ndif = ndiffs(y_train, alpha=0.05, test='adf', max_d=6)
ndif
```

```python
import pmdarima as pm
model = pm.auto_arima(
    y_train, d=1, start_P=0, max_P=3, start_q=0, max_q=3, m=1, seasonal=False, trace=True
)
```

```python
model.fit(y_train)
```

```python
y_pred = model.predict(n_periods=len(y_test))
y_pred
```

```python
plt.figure(figsize=(10, 3))
plt.plot(y_train, label='훈련', color="#9ED1FD")
plt.plot(y_test, label='테스트', color="#FFE1AC")
plt.plot(y_test.index, y_pred, label='예측', color="#FDA7B4")
plt.legend()
plt.show()
```

```python
def one_step():
    pred, conf_int = model.predict(n_periods=1, return_conf_int=True)
    print('예측:', pred.tolist()[0])
    print('에측 하한, 상한:', conf_int[0])
    return pred.tolist()[0], conf_int[0]
```

```python
y_pred=[]
y_lower=[]
y_upper=[]
for test in y_test:
    pred, conf = one_step()
    y_pred.append(pred)
    y_lower.append(conf[0])
    y_upper.append(conf[1])
    #print(pred)
    model.update(test)
```

```python
#티커 종목코드 구하기
import FinanceDataReader as fdr
df =fdr.StockListing('KRX')
names = ['삼성전자', '카카오', 'NAVER', 'LG전자', '현대차']
filt = df['Name'].isin(names)
df = df[filt][['Code', 'Name']]
names = list(df['Name'])
codes = list(df['Code'])
names, codes
```

```python
#1)삼성전자의 start ~ end까지 주식데이터
code = '005930'
start = '2025-01-01'
end = '2025-09-30'
df = fdr.DataReader(code, start, end)
y_train = df['Close']
```

```python
#2)모델생성후 학습
import warnings
warnings.filterwarnings('ignore')

import pmdarima as pm
model = pm.auto_arima(
    y_train, d=1, start_P=0, max_P=3, start_q=0, max_q=3, m=1, seasonal=False, trace=False
)
model.fit(y_train)
```

```python
#3)예측
def one_step():
    pred, conf = model.predict(n_periods=1, return_conf_int=True)
    return pred.tolist()[0], conf[0]
```

```python
#4)예측함수 실행
pred=[]
lower=[]
upper=[]
for y in y_train:
    pred, conf = model.predict(n_periods=1, return_conf_int=True)
    y_pred, conf = one_step()
    pred.append(y_pred.tolist()[0])
    lower.append(conf[0][0])
    upper.append(conf[0][1])
    model.update(y)
```

```python
pred[:5], upper[:5], lower[:5], y_train[:5]
```

```python
#5)결과를 데이터프레임으로 변환
import pandas as pd
df = pd.DataFrame(y_train)
df['예측'] = pred
df['하한'] = lower
df['상한'] = upper
df.head()
```

```python
#6)그래프출력
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

plt.figure(figsize=(10, 3))
df['Close'].plot(label='실제', color="#5DAFF6")
df['예측'].plot(label='예측', color="#F76A7F")
plt.fill_between(df.index, df['하한'], df['상한'], color='k', alpha=0.1)  
plt.grid(True, ls='--', lw=0.5)
plt.title(f'삼성전자 {start}~{end} 주가예측', size=15)
plt.legend()
plt.show()
```
