```python
with open('영어가사.txt', 'r')as file:
    text = file.read()
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud().generate(text)
plt.imshow(wc)
plt.axis('off')
plt.savefig('엘리멘탈.png')
plt.show()
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
with open('애국가.txt', 'r', encoding='utf-8')as file:
    text = file.read()
print(text)
```

```python
okt = Okt()
nouns = okt.nouns(text)
nouns = [noun for noun in nouns if len(noun) > 1]
nouns
```

```python
Counter = Counter(nouns)
Counter
```

```python
import numpy as np
from PIL import Image 
heart_mask = np.array(Image.open('heart_mask.png'))
```

```python
wc = WordCloud(
    width=800,
    height=800,
    font_path='C:/Windows/Fonts/malgun.ttf',  # 또는 사용 가능한 한글 폰트 경로
    scale=2.0,
    max_font_size=250,
    mask=heart_mask,
    background_color="#98D7FE").generate_from_frequencies(Counter)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('애국가.png')
plt.show()
```
