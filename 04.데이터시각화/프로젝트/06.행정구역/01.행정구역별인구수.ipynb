```python
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')
```

```python
#행정구역별 인구수
df = pd.read_excel('행정구역_202401.xlsx', skiprows=1)
df.head(2)
```

```python
df1 = df.copy()
df1.columns = ['행정구역2', '총인구수', '남자인구수', '여자인구수']
df1['행정구역1'] = ''
df1.head(2)
```

```python
df2 = df1.copy()
cols = list(df1.columns)
df2 = df1[[cols[-1]] + cols[0:-1]] #df2 = df1[['행정구역1', '행정구역2'...]]
df2.head(2)
```

```python
sidos = ['특별시', '광역시', '자치시', '경기도', '남도', '북도', '자치도']
for i in range(1, len(df2)):
    for sido in sidos:
        if sido in df2.loc[i, '행정구역2']:
            df2.loc[i, '행정구역1'] = df2.loc[i, '행정구역2']
            df2.loc[i, '행정구역2'] = '소계'
df2
```

```python
for  i in range(2, len(df2)):
    if df2.loc[i, '행정구역1'] == '':
        df2.loc[i, '행정구역1'] = df2.loc[i-1, '행정구역1'] 

df2.loc[0, '행정구역1'] = '전국'
df2.loc[0, '행정구역2'] = '합계'
df2['행정구역2'] = df2['행정구역2'].str.strip()
df2
```

```python
df2.to_csv('행정구역별인구수.csv')
```

```python
plt.rc('font', size='8')
plt.barh(df3['행정구역1'], df3['총인구수']/10000)

xticks = [i for i in range(0, 1501, 150)]
plt.xticks(xticks, [f'{x:,}만' for x in xticks])

for i, p in enumerate(df3['총인구수']/10000):
    plt.text(p+67, i-0.1, f'{p:,.02f}만', ha='center', color = 'midnightblue', size=7)

plt.show()
```
