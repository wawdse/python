```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt
#한글설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
import numpy as np
def graph(name):
    filt = df['이름'] == name
    df2 = df[filt]
    if len(df2)==0:
        return np.nan
    else:
        df3 =  df2.loc[:, ['국어', '과학', '사회']]
        df4 = df3.iloc[0]
        plt.bar(df4.index, df4.values)
        return plt
    
plt=graph('변덕규')
if pd.isna(plt):
    print('없음')
else:
    plt.show()
```

```python
name= df['이름']
height = df['키']
```

```python
plt.barh(name, height)
plt.ylabel('학생이름')
plt.xlabel('키')
plt.xlim(160, 210)
plt.show()
```

```python
plt.figure(figsize=(10, 3))
bar=plt.bar(name, height, color='none', ec='g')
bar[0].set_hatch('/')
plt.ylim(160, 210)
for idx, h in enumerate(height):
    plt.text(idx, h+1, h, ha='center')
plt.show()
```
