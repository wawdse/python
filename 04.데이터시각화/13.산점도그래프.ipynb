```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
import pandas as pd
df = pd.read_csv('data/score.csv')
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
plt.scatter(df['국어'], df['영어'], s=df['학년']*500, c=df['학년'], alpha=0.3 )
plt.colorbar(ticks=[1, 2, 3], orientation='horizontal', label='학년', shrink=0.5)
for idx in range(len(df)):
    x= df.loc[idx, '국어']
    y= df.loc[idx, '영어']
    name = df.loc[idx, '이름']
    plt.text(x, y, name, ha='center', size=8)
    plt.ylim(0, 120)
plt.show()
```

```python
plt.scatter(df['국어'], df['영어'], sizes=df['학년']*500,
            c=df['학년'], cmap='autumn', alpha=0.3) 
plt.xlabel('국어')
plt.ylabel('영어')
plt.colorbar(ticks=[1, 2, 3], orientation='horizontal', label='학년', shrink=0.5)
plt.show()
```
