```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
group = df.groupby('SW특기')
group
```

```python
labels =['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC'] 
```

```python
labels = group.index
values = group.values
labels, values
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=True)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels =['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC'] 
colors = ["#C6FDFB","#A3DAFD","#BADEFB","#85BEF3","#4F78DF","#4659EC"]
```

```python
plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
plt.pie(values, labels = labels, colors=colors, explode=[0.05] * len(values),
        autopct='%.1f%%', startangle=90, counterclock=False)
colors = ["#C6FDFB","#A3DAFD","#BADEFB","#85BEF3","#4F78DF","#4659EC"]
plt.legend(loc=(1.2, 0.5))
plt.show()
```

```python
def custom_autopct(pct):
    if pct <=2:
        return ''
    else:
        return f'{pct:.0f}%'
```

```python
colors = ["#DBFAF9","#ABF3EF","#79D6D3","#47ACA9","#2A8683","#0C5C59"]
plt.pie(values, labels = labels, colors=colors, explode=[0.05] * len(values),
        wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':5},
        autopct='%.1f%%', startangle=90, counterclock=False)
plt.legend(loc=(1.2, 0.5))
plt.show()
```
