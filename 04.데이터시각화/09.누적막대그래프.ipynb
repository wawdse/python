```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['이름'], df['국어'], label='국어', color='#7CFC00')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label='영어', color="#D7ED5B")
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'], label='수학', color="#F1AD79")
plt.ylim(0, 300)
plt.legend()
plt.show()
```

```python
#힉생별(이름), 국어, 영어, 수학점수를 꺽은선 그래프
plt.figure(figsize=(10, 3))
plt.title('학생별 국어, 영어점수', size=15)
plt.plot(df['이름'], df['국어'], label='국어',
         marker='o', ms='10', mec='#4169E1', mew=3, mfc='#FFFACD')
plt.plot(df['이름'], df['영어'], label='영어',
         lw=3, ls='--', color='#40E0D0')
plt.legend(loc=(0.5, -0.2), ncol=2)
plt.ylim(0, 110)
yticks=[y for y in range(0, 101, 50)]
plt.yticks(yticks, [f'{y}점' for y in yticks])
plt.xticks(df['이름'], [x+'님' for x in df['이름']])
plt.xlabel('이름', size=15, loc='right', color='#1E90FF')
plt.ylabel('점수', size=15, loc='top', color='#1E90FF')
plt.show()
```

```python
#학생별(이름) 키 막대그래프 
plt.figure(figsize=(10,3))
plt.bar(df['이름'], df['키'], color='none', hatch='x', ec='#6495ED')
plt.ylim(160, 210)
yticks = [y for y in range(160, 211, 10)]
plt.yticks(yticks, [f'{y}cm' for y in yticks])
xticks=[f'{df.loc[x, "이름"]}\n{df.loc[x, "지원번호"]}' for x in range(len(df))]
plt.xticks(df['이름'], xticks)
plt.show()
```

```python
#학생별 평균점수를 막대그래프
df['평균'] = df.apply(lambda row:row['국어':'사회'].mean(), axis=1)
plt.bar(df['이름'], df['평균'])
for idx, avg in enumerate(df['평균']):
    plt.text(idx, avg, f'{avg:.2f}', ha='center')
plt.show()
```
