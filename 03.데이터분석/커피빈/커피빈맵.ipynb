```python
import requests
import numpy as np
def get_location(address):
    url='https://dapi.kakao.com/v2/local/search/address.json?query='+address
    headers={'Authorization':'KakaoAK 4a8e1acc493134e630bb6e1dd8deb6eb'}
    res = requests.get(url, headers=headers).json()
    if res['documents']:
        x = res['documents'][0]['x']
        y = res['documents'][0]['y']
    else:
        x = np.nan
        y = np.nan
    return x, y
```

```python
x, y=get_location('서울시')
x, y
```

```python
store = pd.read_csv('커피빈매장.csv')
store['X'] = ''
store['Y'] = ''
store.head(2)
```

```python
for i in store.index:
    address = store.loc[i, 'Address']
    x, y = get_location(address)
    store.loc[i, 'X'] = x
    store.loc[i, 'Y'] = y
```

```python
store.isnull().sum()
store.dropna(inplace=True)
```

```python
import folium
import warnings
warnings.filterwarnings(action='ignore')

def get_map(address):
    filt = store['Address'].str.contains(address)
    df = store[filt]
    if len(df.index) == 0:
        print('못찾음')
    else:
        first = df.head(1)
        x = first['X']
        y = first['Y']
        map = folium.Map(location=[y, x], zoom_start=15)
        for idx, row in df.iterrows():
            name=row['Name']
            x = row['X']
            y = row['Y']
            folium.Marker(location=[y, x],
                          popup=folium.Popup(f'[{idx}]{name}', max_width=200),
                          icon=folium.Icon(color='red', icon='star')).add_to(map)
        map.save(f'{address}.html')
get_map('상암')
```
